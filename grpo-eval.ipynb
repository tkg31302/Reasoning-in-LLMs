```python
from datasets import load_dataset
from unsloth import FastLanguageModel
```

```python
model_evaluated = 'unsloth/phi-4-unsloth-bnb-4bit'
hp_max_seq_len = 1024
```

```python
from unsloth import is_bfloat16_supported
import torch
max_seq_length = hp_max_seq_len # Can increase for longer reasoning traces

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_evaluated,
    max_seq_length = max_seq_length,
    load_in_4bit = True, # False for LoRA 16bit
    fast_inference = True, # Enable vLLM fast inference
    gpu_memory_utilization = 0.7, # Reduce if out of memory
)

FastLanguageModel.for_inference(model);
```

```python
SYSTEM_PROMPT = """
Respond in the following format:
<reasoning>
...
</reasoning>
<answer>
...
</answer>
"""

def extract_hash_answer(text: str) -> str | None:
    if "####" not in text:
        return None
    return text.split("####")[1].strip()

# uncomment middle messages for 1-shot prompting
def get_gsm8k_questions(split="test"):
    data = load_dataset('openai/gsm8k', 'main')[split]  # type: ignore
    data = data.map(lambda x: {  # type: ignore
        'prompt': tokenizer.apply_chat_template([
            # {'role': 'system', 'content': SYSTEM_PROMPT},
            # {'role': 'user', 'content': x['question']}
            {'role': 'user', 'content': SYSTEM_PROMPT + '\n' + x['question']}
        ], tokenize=False, add_generation_prompt = True), # continue_final_message=True,
        'answer': extract_hash_answer(x['answer'])
    })  
    return data  
    
dataset = get_gsm8k_questions()
```

```python
from vllm import SamplingParams

sampling_params = SamplingParams(
    temperature = 0.8,
    top_p = 0.95,
    max_tokens = 1024,
)
```

```python
def get_model_answers(x):
  x['output'] = model.fast_generate([x['prompt']], sampling_params = sampling_params, lora_request = None,)[0].outputs[0].text
  return x
```

```python
import pandas as pd
from tqdm import tqdm

tqdm.pandas()

df = pd.DataFrame(dataset)

# Apply function with progress bar
df["output"] = df["prompt"].progress_apply(lambda x: get_model_answers({"prompt": x})["output"])
```
